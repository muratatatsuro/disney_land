```python
import random
import copy
class Human(object):
    def __init__(self,length,width,mutation_prob):
        self.gene_seq=[]
        self.length=length
        self.width=width
        self.mutation_prob=mutation_prob
        #評価
        self.evaluation=0
        
    def generate(self):
        for i in range(self.length):
            g=random.randint(0,self.width)
            self.gene_seq.append(g)
            
    def reset(self):
        self.evaluation=0
        
    def evaluated(self,val):
        """評価値を返す関数"""
        self.evaluation+=val
        
    def instance(self,gene_seq):
        return self
        
    def crossover(self,other):
        """インスタンスを返したい場合は先にそのインスタンスを定義しておく"""
        idx=random.randint(0,self.length-1)
        jdx=random.randint(idx+1,self.length)
        
        pos=[idx,jdx]
        son=Human(self.length,self.width,self.mutation_prob)
        son.setGeneSequence(copy.deepcopy(self.getGeneSequence()))
        
        dau=Human(self.length,self.width,self.mutation_prob)
        dau.setGeneSequence(copy.deepcopy(self.getGeneSequence()))
        
        
        son.setGeneRange(pos,other)
        dau.setGeneRange(pos,self)            
        return son,dau
        
        
        
    def mutate(self):
        for i in range(self.length):
            if random.randint(0,100)/100<self.mutation_prob:
                self.gene_seq[i]=random.randint(0,self.width)
    
    def getEvaluation(self):
        return self.evaluation
        
    def setGeneSequence(self,gene_seq):
        #新しく遺伝配列を定義しなおす
        self.gene_seq=gene_seq
        
    def getGeneSequence(self):
        #遺伝配列を返す関数
        return self.gene_seq
        
    def getGene(self,position):
        return self.gene_seq[position]
    
    def setGene(self,position,value):
        """遺伝配列の中に値を組み込むための関数"""
        self.gene_seq[position]=value 
        
    def setGeneRange(self,pos,par):
        part=par.getGeneSequence()[pos[0]:pos[1]]
        self.gene_seq[pos[0]:pos[1]]=part
```

```python
#world
import functools
import copy
#from human import Human

class World(object):
    def __init__(self):
        return
    
    def evaluate(self,human):
        amount=functools.reduce(lambda x,y:x+y,human.getGeneSequence())
        length=human.length
        #インスタンスコピー
        h=copy.deepcopy(human)
        #human reset
        h.reset()
        h.evaluated(amount/length)
        
        return h
```

```python
#Guest
import random
import copy
from functools import reduce
#from human import Human

"""人間が継承"""
class Guest(Human):
    def __init__(self,length,width,mutation_prob):
        super().__init__(length,width,mutation_prob)
        #ファストパスのチケット格納
        self.tickets=[]
        #時間
        self.clock=0
        self.gotTicket=-121
        
    def reset(self):
        self.evaluation=0
        self.clock=0
        
    def spend(self,time):
        #timeは所要時間
        self.clock+=time
        
    def now(self):
        """現在時刻を返す"""
        return self.clock
    
    def getFastpass(self,fastpass):
        self.tickets.append(fastpass)
        self.gotTicket=self.now()
        
    def showTickets(self):
        """持っているファストパスを返す"""
        return self.tickets
    
    def lastgottickets(self):
        """ファストパスチケットのラスト時間を返す"""
        return self.gotTicket
        
```

```python
#fastpass
class Fastpass(object):
    
    def __init__(self,name):
        self.__name=name
        self.__available=True
        self.__lower=0
        self.__upper=0
        
    def setTime(self,lower):
        self.__lower=lower
        self.__upper=lower+60
        
    def InTime(self,time):
        #間に合うかどうか
        if (time>self.__lower and time<self.__upper):
            return True
        else:
            return False
        
    def enable(self):
        #利用可能か判定する関数
        self.__available=True
    def disable(self):
        self.__available=False
        
    def Available(self):
        """利用可能かを返す関数"""
        return self.__available
    
    def getName(self):
        return self.__name
    
        
```

```python
#from fastpass import Fastpass
class Attraction(object):
    def __init__(self,name,value,latency,fastpass):
        """初期化したものはjsonから取ってきたリスト形式"""
        self.__name=name
        self.__value=value
        self.__latency=latency
        self.__fastpass=fastpass
        
    def __analyzed_tickets(self,tickets,time):
        """所持しているファストパス:tickets
           現在の時刻:time"""
        for ticket in tickets:
            if (ticket.getName()==self.__name and ticket.Available() and ticket.InTime(time)):
                return True
        return False
    
    def ride(self,guest):
        """guestはclass Guestのインスタンス"""
        #乗り物に乗るまでの最低の時間で初期化
        latency=20
        
        now=guest.now()
        
        #fastpassを持っているかどうか
        hasFastpass=self.__analyzed_tickets(guest.showTickets(),now)
        limit=840
        
        if (hasFastpass):
            latency+=10
        elif (now<limit):
            latency+=self.__latency[now//30]
        else:
            latency=0
            
        if ((now+latency)>limit-60):
            value=0
        else:
            value=self.__value
            
        return [latency,value]
    
    def ticket(self,guest):
        #チケットを取るまでの最低の待ち時間での初期化
        latency=10
        #現在時刻
        now=guest.now()
        #最後にファストパスをとった時間
        last=guest.lastgottickets()
        
        #ファストパスインスタンス
        ticket=Fastpass(self.__name)
        
        if ((now-last)>120 and now<840):
            lower=self.__fastpass[now//30]
            if (lower==0):
                ticket.disable()
            ticket.setTime(lower)
        else:
            ticket.disable()
            
        return [latency,ticket]
    
        
        
        
        
```

```python
#disney_world

#from human import Human
#from world import World
#from attraction import Attraction
#from guest import Guest

import functools
import copy

class Disneyworld(World):
    """attractionsはjson形式から取ってくる"""
    def __init__(self,attractions):
        super().__init__()
        #disneyworldで使える形式にする
        self.attractions=[]
        for k,v in attractions.items():
            self.attractions.append(Attraction(k,v['value'],v['latency'],v['fastpass']))
            
    def evaluate(self,human):
        #ゲストインスタンス
        g=Guest(human.length,human.width,human.mutation_prob)
        #遺伝配列を取得して初期化 人間の遺伝配列を取得　gene_seq=human.getGeneSequence()
        g.setGeneSequence(copy.deepcopy(human.getGeneSequence()))
        for s in g.getGeneSequence():
            if (s<len(self.attractions)):
                """乗り物に乗る"""
                result=self.attractions[s].ride(g)
                #時間を計算
                g.spend(result[0])
                #評価値を計算
                g.evaluated(result[1])
            else:
                """ファストパスを取りに行く"""
                result=self.attractions[s-len(self.attractions)-1].ticket(g)
                g.spend(result[0])
                g.getFastpass(result[1])
                
        #園内を出て人間界へようこそ
        h=Human(human.length,human.width,human.mutation_prob)
        #人間の遺伝子配列をセット 園内で遊んできた人間の遺伝子持ってる
        #gene_seq=guest.getGeneSequence()
        h.setGeneSequence(copy.deepcopy(g.getGeneSequence()))
        #人間界で評価
        h.evaluated(g.getEvaluation())
        
        return h
                
```
